```python
#how to create a pivot table 
statewise =pd.pivot_table(data,values =["Confirmed","Deaths","Recovered"],index = "State",aggfunc = max )
```

```python
statewise["Recovery Rate"] = statewise["Recovered"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] =statewise ["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed",ascending = False)
```

```python
# top 10 Active casese
top_10_active_cases =data .groupby(by = "State").max()[["Active_Cases","Date"]].sort_values(by = ["Active_Cases"],ascending = False).reset_index()
fig =plt.figure(figsize=(16,9))
plt.title("Top 10 States with most Active Cases in India",size = 25)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_Cases", x= "State",linewidth = 2, edgecolor= "red")
plt.xlabel("State")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Top states with highest deaths 
top_10_deaths =data.groupby(by = "State").max()[["Deaths","Date"]].sort_values(by = ["Deaths"],ascending = False).reset_index()
fig =plt.figure(figsize=(12,5))
plt.title("Top 10 States with most Deaths Cases",size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x= "State",linewidth = 2, edgecolor= "black")
plt.xlabel("State")
plt.ylabel("Total Deaths Cases")
plt.show()
```
